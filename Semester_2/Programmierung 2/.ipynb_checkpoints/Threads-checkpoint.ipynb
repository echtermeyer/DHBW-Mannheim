```python
from threading import Thread, Lock, RLock
```

```python
thread_lock = Lock()
thread_re_entrang_lock = RLock()
```

```python
num = 5

def set_10():
    global num
    thread_re_entrang_lock.acquire()
    thread_re_entrang_lock.acquire()
    num = 10
    
    thread_re_entrang_lock.release()
    thread_re_entrang_lock.release()
    # thread_lock.release()
    
def set_20():
    global num
    thread_lock.acquire()
    num = 20
```

```python
def do_threading():
    global num
    
    t1 = Thread(target=set_10)
    # t2 = Thread(target=set_20)
    
    t1.start()
    # t2.start()
    
    t1.join()
    # t2.join()
    print(num)
    num = 5
```

```python
for i in range(100):
    do_threading()
```
