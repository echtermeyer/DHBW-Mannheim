```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0, 20, 1)
g = np.array([ 500., 845., 1070., 1130., 1240., 1350., 1470., 1550., 1650., 1650.,
       1715., 1770., 1775., 1815., 1840., 1875., 1875., 1895., 1915.,
       1930.])
```

```python
plt.title("Pasta la vista (Spaghetti)")
plt.xlabel("Kochzeit t in Minuten")
plt.ylabel("Nudelgewicht g in Gramm")

plt.plot(t, g)
plt.grid()
plt.ylim((0, 2000))
```
