```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0, 20, 1)
g = np.array([ 500., 846., 1068., 1130., 1238., 1347., 1471., 1547., 1651., 1648.,
       1715., 1772., 1774., 1814., 1842., 1876., 1875., 1897., 1916.,
       1930.])
```

```python
plt.title("Pasta la vista")
plt.xlabel("Time t in minutes")
plt.ylabel("Weight g in gramm")

plt.plot(t, g)
plt.grid()
plt.ylim((0, 2000))
```
