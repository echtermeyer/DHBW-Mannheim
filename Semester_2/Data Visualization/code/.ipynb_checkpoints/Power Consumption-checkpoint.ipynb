```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from datetime import datetime

plt.style.use("seaborn")
```

```python
df = pd.read_csv("power_consumption/power_consumption_2010.csv", index_col=0)
df.head(3)
```

```python
df["time_dt"] = [datetime.strptime(date, "%d/%m/%Y-%H:%M:%S") for date in [f"{d}-{t}" for d, t in list(zip(df["date"].to_list(), df["time"].to_list()))]]
df["time_dt_hours"] = [d.hour for d in df["time_dt"].to_list()]
df["time_dt_months"] = [d.month for d in df["time_dt"].to_list()]
```

```python
df_times_mean = df.groupby(by="time_dt_hours").mean()
df_times_std = df.groupby(by="time_dt_hours").std()


plt.figure(figsize=(15, 8))
plt.suptitle("Global Power Consumption Tageszeiten Vergleich", size=20)


plt.subplot(121)
plt.title("Mittelwert Daten")
plt.xlabel("Uhrzeit in Stunden")
plt.ylabel("Verbrauch in xxx")

plt.bar(x=df_times_mean.index, height=df_times_mean["globintensity"], color="gray")



plt.subplot(122)
plt.title("Maximum Werte als Caps")
plt.xlabel("Uhrzeit in Stunden")
plt.ylabel("Verbrauch in xxx")

errorbar = df.groupby(by="time_dt_hours").std()["globintensity"]

plt.bar(x=df_times_mean.index, height=df_times_mean["globintensity"], color="gray")
plt.errorbar(x=df_times_mean.index, y=df_times_mean["globintensity"], yerr=errorbar, ls="none", color="gray", capsize=5, capthick=1)

plt.ylim(0)
```

```python
df_months_mean = df.groupby(by="time_dt_months").mean()
df_months_std = df.groupby(by="time_dt_months").std()
months = ["Januar", "Februar", "Mrz", "April", "Mai", "Juni", "Juli", "August", "September", "Oktober", "November"]


plt.figure(figsize=(15, 8))
plt.suptitle("Global Power Consumption Monate Vergleich", size=20)


plt.subplot(121)
plt.title("Mittelwert Daten")
plt.xlabel("Monate")
plt.ylabel("Verbrauch in xxx")

plt.bar(x=df_months_mean.index, height=df_months_mean["globintensity"], color="gray")
plt.xticks(range(1,12), months, rotation=90)



plt.subplot(122)
plt.title("Maximum Werte als Caps")
plt.xlabel("Uhrzeit in Stunden")
plt.ylabel("Verbrauch in xxx")

errorbar = df.groupby(by="time_dt_months").std()["globintensity"]

plt.bar(x=df_months_mean.index, height=df_months_mean["globintensity"], color="gray")
plt.errorbar(x=df_months_mean.index, y=df_months_mean["globintensity"], yerr=errorbar, ls="none", color="gray", capsize=5, capthick=1)
plt.xticks(range(1,12), months, rotation=90)

plt.ylim(0)
```
