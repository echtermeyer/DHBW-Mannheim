```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats

plt.style.use("seaborn")

def test_normal(x):
    k2, p = stats.normaltest(x)
    alpha = 1e-3
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print("It's not a normal distribution")
    else:
        print("It's a normal distribution")
```

# Aufgabe 1

```python
df = pd.read_csv("EvalA/COVID-19_2020.csv")
df[df["country"] == "Europe (total)"].head(1)
```

```python
df["weekly_deaths"] = df["weekly_count"].diff(periods=1)
df_country_only = df[(~df["country_code"].isnull()) & (df["indicator"] == "deaths")].sort_values(by="weekly_deaths", ascending=False).reset_index()

plt.figure(figsize=(15, 8))
plt.suptitle("Aufgabe 1", size=20)


plt.subplot(121)
plt.title("TOP 10: Höchste Anstiege der Todeszahlen")
plt.xlabel("\nLand, Datum des Datensatzes")
plt.ylabel("Anzahl der Todesopfer")
plt.xticks(range(10), 
           [f"{c} - {d}" for c, d in list(zip(df_country_only["country"].iloc[:10], df_country_only["year_week"].iloc[:10]))],
           rotation=90)

plt.bar(x=range(10), height=df_country_only["weekly_deaths"].iloc[:10], color="gray")


plt.subplot(122)
plt.title("TOP 100: Höchste Anstiege der Todeszahlen (quantitativer Vergleich")
plt.xlabel("\nLand, Datum des Datensatzes")
plt.ylabel("Anzahl der Todesopfer")

plt.bar(x=range(100), height=df_country_only["weekly_deaths"].iloc[:100], color="gray")

plt.show()

print(
      "CAPTION: Aufgabe 1"
      f"\nIndien hat in Bezug zur Vorwoche am {df_country_only['year_week'].iloc[0]} "
      f"den höchsten Anstieg mit einer Veränderung von {df_country_only['weekly_deaths'].iloc[0]} erlebt")
```

# Aufgabe 2

```python
df = pd.read_csv("EvalA/COVID-19_2020.csv")
df.head(1)
```

```python
df_c = df[~df["country_code"].isnull()]
df_contis = df_c.groupby(by=["continent", "year_week"]).mean()
df_contis.tail()
```

```python
x = len(df_contis.loc[("Africa", slice(None)), "rate_14_day"])

continents = list(df["continent"].unique())
colors = ["blue", "red", "black", "purple", "green"]

plt.figure(figsize=(15, 8))
plt.title("Aufgabe 2\nDurchschnittliche 14-Tage Rate nach Kontinenten", size=20)

for cont, col in list(zip(continents, colors)):
    plt.plot(range(x), df_contis.loc[(cont, slice(None)), "rate_14_day"].to_list(), c=col, label=cont, marker="o")
    

plt.ylabel("Höhe der 14-Tage-Rate")
plt.xlabel("Zeitlicher Verlauf in Wochen")

plt.xticks(range(x), [date for _, date in df_contis.loc[(cont, slice(None)), "rate_14_day"].index.to_list()],
          rotation=90)
plt.legend()

plt.show()
print("CAPTION: Aufgabe 2\nEuropa war 41x Tagessieger")
```

# Aufgabe 3

```python
df = pd.read_csv("EvalA/COVID-19_2020.csv")
df.tail(1)
```

```python
df_agg = df[(df["country_code"].isnull()) & (df["year_week"] == "2021-22") & (df["indicator"] == "deaths")]
df_agg["anteil"] = df_agg["cumulative_count"] / df_agg["population"] * 100
df_agg = df_agg.sort_values(by="anteil", ascending=False).reset_index()

stds = []
means = []
for continent in df[df["country_code"].isnull()]["country"].unique():
    df_c = df[(df["country"] == continent) & (df["indicator"] == "deaths")]
    df_c["anteil"] = df_c["cumulative_count"] / df_c["population"] * 100
    # test_normal(df_c["anteil"])
    # plt.hist(df_c["anteil"])
    # plt.show()
    stds.append(df_c["anteil"].std())
    means.append(df_c["anteil"].mean())

plt.figure(figsize=(15, 8))

plt.figure(figsize=(15, 8))
plt.title("Aufgabe 3\nAnteil Tode zu Gesamtbevölkerung", size=20)
plt.xlabel("\nKontinent")
plt.ylabel("Anteil in Prozent")

plt.bar(range(len(means)), means, width=0.5, color="gray")
plt.errorbar(df_agg["country"].unique(), y=means, yerr=stds, ls="none", color="gray", capsize=5, capthick=1);

plt.show()

print(
    "CAPTION: Aufgabe 3"
    f"\nAmerica: µ={means[0]}, σ={stds[0]}"
    f"\nEU/EEA: µ={means[1]}, σ={stds[1]}"
    f"\nEurope: µ={means[2]}, σ={stds[2]}"
    f"\nAsia: µ={means[3]}, σ={stds[3]}"
    f"\nAfrica: µ={means[4]}, σ={stds[4]}"
    f"\nOceania: µ={means[5]}, σ={stds[5]}"
)
```
