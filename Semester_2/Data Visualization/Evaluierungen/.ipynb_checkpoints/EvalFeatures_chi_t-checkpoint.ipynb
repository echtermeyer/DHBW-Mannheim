### Statistische Abhängigkeit 1xkategorisch 1xintervall

#### skip neutral days

```python
boing_not0=boing[boing.target_cat!="neutral"]
boing_not0.head()
```

### Student T-Test (independent samples) 

```python
from scipy.stats  import ttest_ind
subset_plus=boing_not0[boing_not0.target_cat=="+"]
subset_minus=boing_not0[boing_not0.target_cat=="-"]
for testvariable in boing_not0.keys():    
    if "cat" not in testvariable:#make T-test only for non-categorial variables        
        tval,pval=ttest_ind(subset_plus[testvariable],subset_minus[testvariable])
        print("testvariable:",testvariable,"tval=",tval,"pval=",pval)
```

### Statistische Abhängigkeit 2xkategorisch

#### X²-Test of categorial variables using contingency tables (provided by Pandas)

```python
from scipy.stats import chi2_contingency

for testvariable in boing_not0.keys():
    if "cat" in testvariable:#make X² test only for categoric variables
    
        contingencyTable=p.crosstab(boing_not0[testvariable],boing_not0.target_cat)

        chi2,pval,dof,expectedFreq=chi2_contingency(contingencyTable)
        print("\n\ntestvariable:",testvariable,"chi2=",chi2,"pval=",'{:f}'.format(pval),"dof=",dof)#show pval in non-scienetific expression
        print(contingencyTable)


```
