```python
def test_normal(x):
    k2, p = stats.normaltest(x)
    alpha = 1e-3
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print("It's not a normal distribution")
    else:
        print("It's a normal distribution")
        
    plt.hist(x)
    plt.show()
```

# Aufgabe 1

```python
df = pd.read_csv("data/crime.csv", encoding='latin1')
df_codes = pd.read_csv("data/offense_codes.csv", encoding="latin1")

df.head(3)
```

```python
df_counts = df.groupby("OFFENSE_CODE_GROUP").count().sort_values("Lat", ascending=False).reset_index()
df_counts.head(3)
```

```python
x_data = df_counts["OFFENSE_CODE_GROUP"].iloc[:3]
y_data = df_counts["INCIDENT_NUMBER"].iloc[:3]

title = "Aufgabe 1 - TOP3 der Verbrechen"
x_label = "\nArt des Verbrechens"
y_label = "Anzahl der Delikte\n"

caption = f"""
CAPTION: Aufgabe 1
{x_data[0]}: {y_data[0]}
{x_data[1]}: {y_data[1]}
{x_data[2]}: {y_data[2]}
"""

plt.figure(figsize=(15, 8))

plt.title(title, size=15)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.xticks(
    rotation=0
)

plt.bar(x=x_data, height=y_data, color="gray")
for i, v in enumerate(y_data.to_list()):
    plt.text(
        list(range(len(x_data)))[i], 
        v + 300, str(v), 
        horizontalalignment="center", 
        verticalalignment="bottom", 
        color="black"
    )

plt.show()
print(caption)
```

# Aufgabe 2

```python
df = pd.read_csv("data/crime.csv", encoding='latin1')
df.shape
```

```python
df = df[df["OFFENSE_CODE_GROUP"] == "Larceny"]
df_district = df.groupby("DISTRICT").count().sort_values("Lat", ascending=False).reset_index()
df_district.head(3)
```

```python
x_data = df_district["DISTRICT"]
y_data = df_district["INCIDENT_NUMBER"]

title = "Aufgabe 2 - Deliktevergleich Bezirke"
x_label = "\nBezirke als Kürzel"
y_label = "Anzahl der Delikte"

caption = f"""
CAPTION: 
{x_data[0]} hat mit {y_data[0]} die meisten Vorfälle
"""

plt.figure(figsize=(15, 8))

plt.title(title)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.xticks(
    rotation=90
)

plt.bar(x=x_data, height=y_data, color="gray")


plt.show()
print(caption)
```
